```python
import pandas as pd

url_dados = 'https://github.com/EloizioHMD/imersaodados3/raw/main/dados/dados_experimentos.zip'
dados = pd.read_csv(url_dados, compression= "zip")
dados
```

```python
dados_mod = dados.copy()
dados_mod.columns = [column.replace("-", "") for column in dados_mod.columns]
dados_mod
```

```python
mapa = {'droga' : 'composto'}
dados_mod.rename(columns=mapa, inplace=True)
dados_mod
```

### Desafio 2
Melhorar a visualizaÃ§Ã£o alterando tamanho da fonte, para deixar os grÃ¡ficos mais atrativos.

```python
plt.figure(figsize=[8,6])
ax = sns.countplot(x = 'composto', data=dados_mod.query('composto in @cod_compostos'), order=cod_compostos[0:-1])
ax.set_title('Top 5 Compostos', fontsize= 24, fontweight = 'bold')
ax.set_ylabel('frequÃªncia', fontsize=14, fontweight = 'bold')
ax.set_xlabel('compostos', fontsize=14, fontweight = 'bold')
for p in ax.patches:
    percentage = '{:,.0f}'.format(p.get_height())
    width, height = p.get_width(), p.get_height()
    x = p.get_x()+width/3
    y = p.get_y()+height+20
    ax.annotate(percentage,(x,y))
plt.show()
```

### Desafio 3
Plotar os histogramas com seaborn

```python
fig, ax = plt.subplots(1,4,figsize=(30,5))
plt.suptitle('GrÃ¡ficos estatÃ­stico do `.describe()`', fontsize=16, fontweight='bold')
sns.histplot(data=dados_mod, x=(dados_mod.loc[:,'g0':'g771'].describe().T['mean']), ax = ax[0], kde=True, bins=30).set_title('MÃ©dia', fontsize=14)
sns.histplot(data=dados_mod, x=(dados_mod.loc[:,'g0':'g771'].describe().T['std']), ax = ax[1], kde=True, bins=30).set_title('Desvio PadrÃ£o', fontsize=14)
sns.histplot(data=dados_mod, x=(dados_mod.loc[:,'g0':'g771'].describe().T['max']), ax = ax[2], kde=True, bins=30).set_title('MÃ¡ximo', fontsize=14)
sns.histplot(data=dados_mod, x=(dados_mod.loc[:,'g0':'g771'].describe().T['min']), ax = ax[3], kde=True, bins=30).set_title('MÃ­nimo', fontsize=14)
plt.show()
```

As estatÃ­sticas descritivas incluem aquelas que resumem a tendÃªncia central, a dispersÃ£o e a forma da distribuiÃ§Ã£o de um conjunto de dados, excluindo os valores NaN.

Analisa sÃ©ries numÃ©ricas e de objetos, bem como conjuntos de colunas DataFrame de tipos de dados mistos. A saÃ­da irÃ¡ variar dependendo do que Ã© fornecido. Consulte as notas abaixo para obter mais detalhes.
### NumÃ©ricos
Para dados numÃ©ricos, o Ã­ndice do resultado incluirÃ¡ contagem, mÃ©dia, padrÃ£o, mÃ­nimo, mÃ¡ximo, bem como percentis inferiores, 50 e superiores. Por padrÃ£o, o percentil inferior Ã© 25 e o percentil superior Ã© 75. O percentil 50 Ã© igual Ã  mediana.
### Objetos
Para dados de objeto (por exemplo, strings ou carimbos de data / hora), o Ã­ndice do resultado incluirÃ¡ count, unique, top e freq. O topo Ã© o valor mais comum. O freq Ã© a frequÃªncia do valor mais comum. Os carimbos de data / hora tambÃ©m incluem o primeiro e o Ãºltimo itens.

Se vÃ¡rios valores de objeto tiverem a contagem mais alta, a contagem e os resultados principais serÃ£o escolhidos arbitrariamente entre aqueles com a contagem mais alta.
### Dataframe
Para tipos de dados mistos fornecidos por meio de um DataFrame, o padrÃ£o Ã© retornar apenas uma anÃ¡lise de colunas numÃ©ricas. Se o dataframe consistir apenas em objetos e dados categÃ³ricos sem quaisquer colunas numÃ©ricas, o padrÃ£o Ã© retornar uma anÃ¡lise do objeto e das colunas categÃ³ricas. Se include = 'all' for fornecido como uma opÃ§Ã£o, o resultado incluirÃ¡ uma uniÃ£o de atributos de cada tipo.

Os parÃ¢metros de inclusÃ£o e exclusÃ£o podem ser usados â€‹â€‹para limitar quais colunas em um DataFrame sÃ£o analisadas para a saÃ­da. Os parÃ¢metros sÃ£o ignorados ao analisar uma sÃ©rie.

Para mais informaÃ§Ãµes consultar [pandas.DataFrame.describe](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.describe.html).
