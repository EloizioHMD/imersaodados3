```python
import pandas as pd

url_dados = 'https://github.com/EloizioHMD/imersaodados3/raw/main/dados/dados_experimentos.zip'
dados = pd.read_csv(url_dados, compression= "zip")
dados
```

```python
dados_mod = dados.copy()
dados_mod.columns = [column.replace("-", "") for column in dados_mod.columns]
dados_mod
```

```python
mapa = {'droga' : 'composto'}
dados_mod.rename(columns=mapa, inplace=True)
dados_mod
```

Trabalhando com Gráficos do Seaborn. Para mais informação consultar a [documentação no site](https://seaborn.pydata.org/api.html) do projeto. 

O gráfico possuí muitas variáveis, então fica bem ruim de visualizar e obter alguma compreensão por gráficos. Então vamos realizar o mesmo gráfico com uma filtragem para os 5 valores mais frequentes.

```python
cod_compostos =  dados_mod['composto'].value_counts().index[0:5]
cod_compostos
```

### Desafio 2
Melhorar a visualização alterando tamanho da fonte, para deixar os gráficos mais atrativos.

```python
plt.figure(figsize=[8,6])
ax = sns.countplot(x = 'composto', data=dados_mod.query('composto in @cod_compostos'), order=cod_compostos[0:-1])
ax.set_title('Top 5 Compostos', fontsize= 24, fontweight = 'bold')
ax.set_ylabel('frequência', fontsize=14, fontweight = 'bold')
ax.set_xlabel('compostos', fontsize=14, fontweight = 'bold')
for p in ax.patches:
    percentage = '{:,.0f}'.format(p.get_height())
    width, height = p.get_width(), p.get_height()
    x = p.get_x()+width/3
    y = p.get_y()+height+20
    ax.annotate(percentage,(x,y))
plt.show()
```

### Desafio 3
Plotar os histogramas com seaborn

```python
fig, ax = plt.subplots(1,4,figsize=(30,5))
plt.suptitle('Gráficos estatístico do `.describe()`', fontsize=16, fontweight='bold')
sns.histplot(data=dados_mod, x=(dados_mod.loc[:,'g0':'g771'].describe().T['mean']), ax = ax[0], kde=True, bins=30).set_title('Média', fontsize=14)
sns.histplot(data=dados_mod, x=(dados_mod.loc[:,'g0':'g771'].describe().T['std']), ax = ax[1], kde=True, bins=30).set_title('Desvio Padrão', fontsize=14)
sns.histplot(data=dados_mod, x=(dados_mod.loc[:,'g0':'g771'].describe().T['max']), ax = ax[2], kde=True, bins=30).set_title('Máximo', fontsize=14)
sns.histplot(data=dados_mod, x=(dados_mod.loc[:,'g0':'g771'].describe().T['min']), ax = ax[3], kde=True, bins=30).set_title('Mínimo', fontsize=14)
plt.show()
```

As estatísticas descritivas incluem aquelas que resumem a tendência central, a dispersão e a forma da distribuição de um conjunto de dados, excluindo os valores NaN.

Analisa séries numéricas e de objetos, bem como conjuntos de colunas DataFrame de tipos de dados mistos. A saída irá variar dependendo do que é fornecido. Consulte as notas abaixo para obter mais detalhes.
### Numéricos
Para dados numéricos, o índice do resultado incluirá contagem, média, padrão, mínimo, máximo, bem como percentis inferiores, 50 e superiores. Por padrão, o percentil inferior é 25 e o percentil superior é 75. O percentil 50 é igual à mediana.
### Objetos
Para dados de objeto (por exemplo, strings ou carimbos de data / hora), o índice do resultado incluirá count, unique, top e freq. O topo é o valor mais comum. O freq é a frequência do valor mais comum. Os carimbos de data / hora também incluem o primeiro e o último itens.

Se vários valores de objeto tiverem a contagem mais alta, a contagem e os resultados principais serão escolhidos arbitrariamente entre aqueles com a contagem mais alta.
### Dataframe
Para tipos de dados mistos fornecidos por meio de um DataFrame, o padrão é retornar apenas uma análise de colunas numéricas. Se o dataframe consistir apenas em objetos e dados categóricos sem quaisquer colunas numéricas, o padrão é retornar uma análise do objeto e das colunas categóricas. Se include = 'all' for fornecido como uma opção, o resultado incluirá uma união de atributos de cada tipo.

Os parâmetros de inclusão e exclusão podem ser usados ​​para limitar quais colunas em um DataFrame são analisadas para a saída. Os parâmetros são ignorados ao analisar uma série.

Para mais informações consultar [pandas.DataFrame.describe](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.describe.html).
